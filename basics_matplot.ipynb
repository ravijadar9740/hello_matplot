```python
# importing the 'numpy' module for mathematical expressions
import numpy as np
# extracting pyplot from the matplotlib module
import matplotlib.pyplot as plt
```

```python
# declaring x and y as dataset for the plots
x=[ x for x in range(2,8)]
y=[ y for y in range(5,11)]
```

```python
# check for the length of data of x & y for they should same  
len(x)
len(y)
```

```python
# accessing the plots from matplotlib.pyploy as we declared them as 'plt'
# 'scatter' is the command for framing dataset in the form of points  
plt.scatter(x,y,c="y")# 'c' implies colour
# setting the axes limits
plt.ylim(1,15)
plt.xlim(1,10)
```

```python
# generating the random dataset for x & y 
x=np.random.randint(5,90,9)
y=np.random.normal(200,7,9)
```

```python
# just extending by giving nameas 'label' for dataset as label and s'' implies for size
plt.scatter(x,y,c='r',label="ravi's data",s=10)
plt.legend()# display the label
plt.title(" ravi's point graph ")# for title of the graph
# naming for axes
plt.xlabel(" x axis")# for x axis 
plt.ylabel(" y axis")# for y axis
```

```python
# 'plot' command displays the dataset in the joint form
plt.plot(x,y,c='r',label="ravi's data",lw=0.7)# 'lw' stands for line width 
plt.legend()
plt.title(" ravi's line graph ")
plt.xlabel(" x axis")
plt.ylabel(" y axis")
```

```python
# declaring the dataset for trignnometric functions as we ae accessing the pi in dataset
# 'linspace' command works as taking values like (start,end,no_of_points)
x=np.linspace(-3*np.pi,+5*np.pi,100)
y=np.sin(x)# assigning the x as sin functions in y
```

```python
# performing operation for sin wave
plt.plot(x,y,c='r',label="ravi's data",lw=0.7)
plt.legend()
plt.title(" ravi's sin wave")
plt.xlabel(" x axis")
plt.ylabel(" y axis")
```

```python
# same for tan 
y=np.tan(x)
```

```python
plt.plot(x,y,c='r',label="ravi's data",lw=0.7)
plt.legend()
plt.title(" ravi's tan wave")
plt.xlabel(" x axis")
plt.ylabel(" y axis")
```

```python
# declaring dataset for bar graph
x1=["a","b","c","d","e"]
y1=[76,87,98,59,59]
```

```python
plt.bar(x1,y1,label="ravi's data")#'bar' command used for getting bar graph
plt.legend()
plt.title(" ravi's bar graph")
plt.xlabel(" x axis")
plt.ylabel(" y axis")
```

```python
# performing for histograph
plt.hist(y1,label="ravi's data")# 'hist' command is used for getting histograph
plt.legend()
plt.title(" ravi's hist-graph")
plt.xlabel(" x axis")
plt.ylabel(" y axis")
```

```python
# generating random dataset for histograph
x2=np.random.normal(100,8,200)
```

```python
# perorming histograph for generated random dataset
plt.hist(x2,label="ravi's data")
plt.legend()
plt.title(" ravi's hist-graph")
plt.xlabel(" x axis")
plt.ylabel(" y axis")
```

```python
x3=np.random.normal(100,8,200)
```

```python
# here we using command 'bin' for range of values & command 'width' for bar width
plt.hist(x3,label="ravi's data",bins=20,width=1)
plt.legend()
plt.title(" ravi's hist-graph")
plt.xlabel(" x axis")
plt.ylabel(" y axis")
```

```python
# declaring the dataset for proceeding with piechart
p=["a","b","c","d","e"]
q=[1,2,3,4,5]
```

```python
plt.pie(q,labels=p)# command 'pie' used here for feching piechart
plt.legend()
plt.title("pie chart")
```

```python
plt.pie(q,labels=p,autopct="%.3f")# here the command 'autopict' used to display the ratio of each part 
plt.legend()
plt.title("pie chart")
```

```python
ex=[0,0,0,0.1,0]# declaring for 'explode' command
```

```python
plt.pie(q,labels=p,autopct="%.3f",explode=ex)# command 'explode' used for spacing the desired part from the piechart
plt.legend()
plt.title("pie chart")
```
